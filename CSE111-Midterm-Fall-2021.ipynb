```python
# Write you code here
class Vehicle:
    def __init__(self,place,vehi,orga,dic):
        self.place=place
        self.vehi=vehi
        self.orga=orga
        self.dic=dic
        self.count=0
        self.info=[]
    def addPassenger(self,*info):
        for obj in info:
            if obj.city in self.place:
                self.name=obj.name
                print(f'{obj.name} has taken the {self.vehi}.')
                self.count+=1
                self.info.append(obj.name)
                self.info.append(obj.city)
            else:
                print(f"This is an Inter-City {self.vehi}. {obj.name} cannot take the {self.vehi}.")
    def status(self):
        fee=0
        for key,val in self.dic.items():
            if key==self.orga:
                fee=val
        if fee==0:
            fee=90
        print(f"Details of {self.place} {self.vehi} of {self.orga}:\nTotal Passengers: {self.count}")
        for i in range(0,len(self.info),2):
            print(f"Passenger name : {self.info[i]}, Destination: {self.info[i+1]}, Ticket Price: {fee}")
        
class Passenger:
    def __init__(self,name,city='None'):
        self.name=name
        if city=='None':
            print("Destination cannot be empty!")
        else:
            self.city=city
    def setDestination(self,city):
        self.city=city
        print(f"Destination of {self.name} is updated to {self.city}")
    def calculateFare(self,obj):
        print(f"Calculating fare for {obj.name}")
        self.calc=0
        for key,val in obj.dic.items():
            if key==obj.orga:
                self.calc+=val
        
        
        


#=================================================
#Do not change the codes below

fareChart = {'SR Travels': 40, 'ENA': 50, 'Pathao': 200, 'Uber': 
250, 'Hanif Paribahan': 45, 'Others': 90 }
v1 = Vehicle("Dhaka-123", "Bus", 'SR Travels', fareChart)
p1 = Passenger("Sam", "Dhaka")
print("1.===========================")
v1.addPassenger(p1)
print("2.===========================")
p1.calculateFare(v1)
print("3.===========================")
p2 = Passenger("John", "Dhaka")
p3 = Passenger("Bran")
print("4.===========================")
p3.setDestination("Chittagong")
print("5.===========================")
v1.addPassenger(p2, p3)
print("6.===========================")
p2.calculateFare(v1)
print("7.===========================")
p4 = Passenger("Mark", "Dhaka")
p5 = Passenger("Jack", "Khulna")
v1.addPassenger(p4, p5)
print("8.===========================")
p4.calculateFare(v1)
print("===========================")
v1.status()
v2 = Vehicle("Chittagong-798", "Car", "Uber", fareChart)
print("9.===========================")
v2.addPassenger(p1)
v2.addPassenger(p3)
print("10.===========================")
p3.calculateFare(v2)
print("11.===========================")
v2.status()
v3 = Vehicle("Khulna-997", "Car", "Obhai", fareChart)
v3.addPassenger(p5)
print("12.===========================")
p5.calculateFare(v3)
print("13.===========================")
v3.status()
```
